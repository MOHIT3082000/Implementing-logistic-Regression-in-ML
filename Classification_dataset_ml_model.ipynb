```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler
```

```python
x = df.drop(columns=['Preference'])
y = df['Preference']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3, random_state=42)
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
logreg = LogisticRegression(solver='liblinear')
param_grid = {
    'C':[0.01,0.1,1,10,100],
    'penalty':['l1','l2'],
    'max_iter':[100,200,300]
}
grid_search = GridSearchCV(logreg,param_grid, cv=5, scoring='accuracy', verbose=1, n_jobs=-1)
grid_search.fit(x_train, y_train)
```

```python
best_params = grid_search.best_params_
print("Best Parameters:", best_params)
print("Best Training Accuracy:", grid_search.best_score_)
```

```python
best_logreg = grid_search.best_estimator_
y_pred = best_logreg.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
print("Model Performance:")
print(f"Accuracy:{accuracy:.2f}")
print(f"Precision:{precision:.2f}")
print(f"Recall:{recall:.2f}")
print(f"F1 Score:{f1:.2f}")
```

```python
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier(random_state = 42)
```

```python
param_grid_dt = {
    'criterion':['gini','entropy'],
    'max_depth':[None,10,20,30,50],
    'min_samples_split':[2,10,20],
    'min_samples_leaf':[1,5,10]
}
```

```python
grid_search_dt = GridSearchCV(dtree, param_grid_dt,cv=5, scoring='accuracy',verbose=1, n_jobs=-1)
grid_search_dt.fit(x_train,y_train)
```

```python
best_params_dt = grid_search_dt.best_params_
print("Best Parameters:", best_params_dt)
print("Best Training Accuracy:", grid_search_dt.best_score_)
```

```python
best_dtree = grid_search_dt.best_estimator_
```

```python
y_pred_dt = best_dtree.predict(x_test)
```

```python
accuracy = accuracy_score(y_test,y_pred_dt)
precision = precision_score(y_test,y_pred_dt)
recall = recall_score(y_test,y_pred_dt)
f1 = f1_score(y_test,y_pred_dt)
```

```python
print("Model Performance:")
print(f"Accuracy:{accuracy:.2f}")
print(f"Precision:{precision:.2f}")
print(f"Recall:{recall:.2f}")
print(f"F1 score:{f1:.2f}")
```

```python
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_dt))
print("Classification Report:\n", classification_report(y_test, y_pred_dt))
```
